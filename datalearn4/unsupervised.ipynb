```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
cancer=load_breast_cancer()
x_train,x_test,y_train,y_test=train_test_split(cancer.data,cancer.target,random_state=1)
print(x_train.shape)
print(x_test.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
We then fit the scaler using fit method applied to training data.For
MinMaxScaler the fit method computes the minimum and maximum value of each feature
on the training set. In contrast to lassifer and regressors. the scaler is only
proviided with data (x_train) when fit is called y_train is not used.
```

```python
scaler.fit(x_train) #creates a sk learn scaler. 
```

```python
To apply the transformation that we just learned that is actually scale the training data we use transform
method of scaler. The transform method is used in scikit learn whenever a model retursn
a new representation of the data.
```

```python
x_train_scaled=scaler.transform(x_train) #transform returns a new data. allow it to turn into new data. 
```

```python
print("original shape : {}".format(x_train.shape))
print("transformed shape: {}".format(x_train_scaled.shape))
print("pre feature minimum before scaling:\n {}".format(x_train.min(axis=0)))
print("pre feature maximum before scaling: \n {}".format(x_train.max(axis=0)))
print("pre feature maximum ater scaling :\n {}".format(x_train_scaled.min(axis=0)))
print("pre feature maximum after scaling:\n {}".format(x_train_scaled.max(axis=0)))
```

```python
THe transformed data has the same shape as the original data, features are simply shifted
and scaled .You can see that all of the features are now between 0 and 1 as desired.
To apply the SVM to the scaled data,we also need to transform the test set. This is again 
done by calling the transform method this tiime on x-test
```

```python

x_test_scaled=scaler.transform(x_test)
print("pre feature minimum after scaling :\n {}".format(x_test_scaled.min(axis=0)))
print("per feature maximum after scaling :\n{}".format(x_test_scaled.max(axis=0)))
```

```python
Maybe somewhat suprisingly you can see that for test set, after scaling the minimum and maximum are not 
0 and1. some of the features are even outside the 0-1 range. THe explanation is that MinMaxScaler ( all
other scalers)
always applies exactly the same transformation to the training and testset. This mean the 
transform method always subtracts the training set minimum and divides y the training set range
which might be differnet from minimum and range for the testss set.
It is important to apply exaclty the same tranformation to the training set and test sett for the
supervised  model to work on the test set.
```

```python

from sklearn.model_selection import train_test_split
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import mglearn
x,_ = make_blobs(n_samples=50,centers=5,random_state=4, cluster_std=2) #its std for one cluster 
#make blob is a random data generator. avaliable in sklearn. center 5. 5 times happening is the center
#random state is shuffling. 
x_train,x_test=train_test_split(x,random_state=5,test_size=0.1)
fig,axes=plt.subplots(1,3,figsize=(13,4))#make blob. random data generator 
#avaliable from sklearn library 
axes[0].scatter(x_train[:,0],x_train[:,1],c=mglearn.cm2(0),label="Training set",s=60)
axes[0].scatter(x_test[:,0],x_test[:,1],marker='^',c=mglearn.cm2(1),label="Test set", s=60) #
axes[0].legend(loc="upper left")
axes[0].set_title("Original Data")

scaler=MinMaxScaler()
scaler.fit(x_train)

x_train_scaled=scaler.transform(x_train)
x_test_scaled=scaler.transform(x_test)

axes[1].scatter(x_train_scaled[:,0],x_train_scaled[:,1],c=mglearn.cm2(1),label="Test set",s=60)
axes[1].scatter(x_test_scaled[:,0],x_test_saled[:,1],marker='^',c=mglearn.cm2(1),label="Test set",s=60)
axes[1].set_title("Scaled Data")
test_scaler=MinMaxScaler()
test_scaler.fit(x_test)

x_test_scaled_badly=test_scaler.transform(x_test)
axes[2].scatter(x_train_scaled[:,0],x_train_scaled[:,1],c=mglearn.cm2(0),label="training set",s=60)
axes[2].scatter(x_test_scaled_badly[:,0],x_test_scaled_badly[:,1],marker='^',c=mglearn.cm2(1),label="test set",s=60)
axes[2].set_title("improperly scaled data")
for ax in axes:
    ax.set_xlabel("Feature 0")
    ax.set_ylabel("Featrure 1")
fig.tight_layout()
```
