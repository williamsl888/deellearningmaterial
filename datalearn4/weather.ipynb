```python
Multi label Clustering
We need to calculate cluster centers. That way for given new sample we can measure distance from
each cluster Using some distance threshold we will be able to assign new data to one or more clusters.
```

```python
fig, ax = plt.subplots(1,2,figsize=(15,5))
events_prepared.sum().sort_values(ascending=False).plot.bar(ax=ax[0],title='Real Events that happened')
resultDf.iloc[:,0].value_counts().plot.bar(ax=ax[1],title='Histogram obtained from agglomerative clustering')
#side by side graph 
```

```python
event_name_ordered=events_prepared.sum().sort_values(ascending=False).index 
clusters_ordered=resultDf.iloc[:,0].value_counts().index
cluster_category_mapping={}
for i in range(clustered_count):
    cluster_category_mapping.update({clusters_ordered[i]:event_name_ordered[i]}) #this gives you the values. running
    #a loop for each cluster. 
cluster_category_mapping
```

```python
resultDf
```

```python
#Find clusters centroids to classify test data
#We need to ave a way to classify new data on already trained clusters
#we can do this by calculating clusters center and mesure distance from each center
cluster_centers_mapping={}
for key in cluster_category_mapping:
    cluster_indices=resultDf.loc[resultDf[0]==key].index #1 3 ,3 ,2 is the key. result. resultdf is the 
    #machine learning predicted values. [predicted 400 row in 0. key shows the rows. 
    cluster_data=X_train.iloc[cluster_indices] #so 1040 o 0 - 1044 are the indexs 
    mean=cluster_data.mean(axis=0).values #mean. all the rows of the keys we will get a mean 
    cluster_centers_mapping.update({key:mean}) #can be like 200, 3000 
cluster_centers_mapping 
# or 2,1,3,4 the mean will be these. 
```

```python
#We now need to calculate distance between these centroids and our data points
def get_distances_from_cluster(data_frame): #making new data and calling it data_Frame. 
    cluster_distance=np.zeros((data_frame.shape[0],clustered_count)) #0 in 1200 rows. shape of 0 is number 
    #of rows in the dataframe. clustered count is 5 numbers like thunder,rain. 
    for i in range(data_frame.shape[0]): #runs 1200 rows. runs that amount of time. 
        for key in cluster_category_mapping:
            dist=np.linalg.norm(data_frame.iloc[[i]].values[0]-cluster_centers_mapping[key])
            #np.linear is the function. norm. used to calculate the distances. training and mean and subtract it. 
            #dfrom our original data from the mean. 7 calculate distance then sqwuare and square it. 
            cluster_distance[i,key]=dist #distance is put in there. 
    column_names=[cluster_category_mapping[k] for k in cluster_category_mapping] #1:rain , 2:thunderstorm. this
    #is our k . rain goes into the column name. it then it loops 
    return pd.DataFrame(cluster_distance, index=data_frame.index,columns=column_names) #our distance.
#we add the rows then square then square root, and we do that to all. data_frame - is our training data. 
#70 is the x_train.head which is our data_frame.index. 
distanceDf=get_distances_from_cluster(X_train) #from line number 2. 80 percent training data. 
distanceDf.head()
#cluster mpapping - train data and minus the first 4 column, then rearrange it into rain, fog. etc 
#we are using tempavgf to predict rain. 
```

```python
Column with minimum distance is our classification, but to have ability to classify a record to both clusters
we can add some tolerance For example
Data points belongs to cluster if distancer from cluster is smaller than minimum distance *1.2
```

```python
def classify_events(distances_dataFrame): 
    return distances_dataFrame.apply(lambda x:x<x.min()*1.02,axis=1) #if the value is so minimum. we 
#multiply it by 1.02. 
clasification_result=classify_events(distanceDf) #distance data into number 1. upscaling multiple with 1.02 
#we sorted it.
#30 contains minimum value. 30 becomes 0 reindexing, minimum value becomes first one now 
X_train_col_ordered=clasification_result.reindex(sorted(clasification_result.columns),axis=1) 
#find the minimum value. makes our data frame shorted one. actually shows you the columns of x train. 
y_train_col_ordered=y_train.reindex(sorted(y_train.columns),axis=1)
```

```python
def check_accuracy(X,y):
    comparison=X==y #train vs test 
    val_counts=comparison.all(axis=1).value_counts()
    percentageCorrect=val_counts.at[True]/X.shape[0]*100 #convert it to percentage 200/300 is 2/3. x.shape 
    #is total number of rows. 
    return percentageCorrect
```

```python
check_accuracy(X_train_col_ordered,y_train_col_ordered) #check percentage, how many are correct 
```
